## EXTRACT

Como a API disponível para o desafio parou de funcionar, eu gerei na IA três usuários diferentes dentro de uma planilha do excel 

```python
import pandas as pd

# Importando os valores de uma planilha -----------------

df = pd.read_excel('USUÁRIOS.xlsx')

#definindo os usuários ---------------------------------

usuarios = df['name'].tolist()

# print para mostrar o que tem dentro da planilha ----

print(df,'\n')
print('Os usuários são:','\n',usuarios)
```

## TRANSFORM

Após extrair os dados do excel, eu conecto com a IA da GROQ, criando respostas para cada um dos usuários presentes no excel

Infelizmente eu não consegui conectar ao ChatGPT por ser pago, mas o do GROQ tem sua versão grátis

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv, find_dotenv

# Essa é a senha da API que está no meu computador em forma de arquivo ----------- 
load_dotenv(find_dotenv())

# O prompt usado -----------------------------------------------------------------

template = """
Crie uma mensagem curta (máx. 70 caracteres), carinhosa e motivadora
sobre a importância de investir.
A mensagem deve soar natural, positiva e diferente das anteriores.
Use o nome da pessoa.

Nome: {nome}
"""
prompt = PromptTemplate.from_template(template)

# Chamando o chat da GROQ -------------------------------------------------------

chat = ChatGroq(model="llama-3.3-70b-versatile")
parser = StrOutputParser()
chain = prompt | chat | parser

# Lista para armazenar mensagens (por uma questão de ser mais fácil resolvi armazenar em lista) ---

mensagem_personalizada = []

for pessoa in usuarios:
    texto = chain.invoke({"nome": pessoa})
    mensagem_personalizada.append(texto)
    print(texto)
```

## LOAD

Por fim, como a API da DIO não possui mais acesso, irei fazer ele criar um excel simples e salvar.

```python
# Criando um dataframe para poder exportar as mensagens ----------------

df_personalizado = pd.DataFrame({
    "id": df['id'],
    "nome": df['name'] ,
    "mensagem": mensagem_personalizada})

print(df_personalizado)

# Gerando o excel e salvando no PC -------------------------------------

df_personalizado.to_excel("ResultadoFinal.xlsx", index=False)
```
